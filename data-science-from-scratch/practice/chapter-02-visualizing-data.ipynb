# Chapter 2: Visualizing Data

### Matplotlib

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.5, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(years, gdp, color="green", marker="o", linestyle="solid")

plt.title("Nominal GDB")

plt.ylabel("Billions of $")
plt.xlabel("Year")

plt.show()
```

## Bar Charts

Bar charts are good for displaying a *discrete* set of items

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Ghandi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

#xs = [i + 0.1 for i, _ in enumerate(movies)]

plt.bar(movies, num_oscars)

plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")

#plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

plt.show()
```

Bar charts are also good for histograms of bucketed numeric values in order to show how values are distributed.

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 100 *10
histogram = Counter(decile(grade) for grade in grades)
```
